```python
from arena_api.system import system
# 查找并连接设备
devices = system.create_device()
if len(devices) == 0:
   print("未检测到任何设备，请检查连接！")
else:
   device = devices[0]
   print(f"已成功连接设备: {device}")
```

```python
nodemap = device.nodemap
tl_stream_nodemap = device.tl_stream_nodemap

	# Enable stream auto negotiate packet size
tl_stream_nodemap['StreamAutoNegotiatePacketSize'].value = True

	# Enable stream packet resend
tl_stream_nodemap['StreamPacketResendEnable'].value = True
```

```python
nodemap['ExposureAuto'].value = 'Off'
exposure_time = 1000.0
nodemap['ExposureTime'].value = exposure_time
```

```python
nodemap['ExposureTime'].min
```

```python
frame_rate = nodemap['AcquisitionFrameRate'].value
print(f"当前采集帧率: {frame_rate} fps")
```

```python
import ctypes
import numpy as np
num_of_buffers = 1
device.start_stream(num_of_buffers)
print("开始采集图像...")
    # 从缓冲区获取一帧图像，超时时间设为2000毫秒
```

```python
buffer = device.get_buffer()
    
    # 从缓冲区创建图像对象并保存
pdata_as16 = ctypes.cast(buffer.pdata, ctypes.POINTER(ctypes.c_ushort))

nparray_reshaped = np.ctypeslib.as_array(
        pdata_as16,
        (buffer.height, buffer.width))
    
print(nparray_reshaped.shape)
    # 归还缓冲区
device.requeue_buffer(buffer)
```

```python
device.stop_stream()
```

```python
from pyvcam import pvc 
from pyvcam.camera import Camera   

pvc.init_pvcam()                   # Initialize PVCAM 
cam = next(Camera.detect_camera()) # Use generator to find first camera
cam.open()                         # Open the camera
```

```python
# A camera object named cam has already been created
frame = cam.get_frame(exp_time=20)
print("First five pixels of frame: {}, {}, {}, {}, {}".format(*frame[:5]))
```

```python
cam.close()
pvc.uninit_pvcam()
```
